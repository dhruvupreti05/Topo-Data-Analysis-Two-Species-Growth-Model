```python
import numpy as np
import matplotlib.pyplot as plt

L = 200
x = np.arange(L)
h = np.cumsum(np.random.randn(L))

plt.plot(x,h)
plt.xlabel('position'); plt.ylabel('height')
plt.show()
```

```python
import gudhi

L = 200
state = np.zeros(L, dtype=int)
state[:L//2] = 1  # half full
np.random.shuffle(state)

T = 10000
for t in range(T):
    i = np.random.randint(L-1)
    if state[i] == 1 and state[i+1] == 0:
        state[i], state[i+1] = 0, 1

# Build the height function from the particle configuration
h = np.cumsum(2*state - 1)  # each particle = +1 step, each hole = -1
plt.plot(h)
plt.show()
```
